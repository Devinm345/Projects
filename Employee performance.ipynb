```python
# 17. Create a new column Experience Level with values "Junior", "Mid", or "Senior" based on the Experience. 
# Convert 'Experience' column to numeric, coercing errors to NaN
df['Experience'] = pd.to_numeric(df['Experience'], errors='coerce')

# Now apply the classification based on Experience
conditions = [
    (df['Experience'] < 5),   # Junior: Less than 5 years of experience
    (df['Experience'] < 10),  # Mid: Less than 10 years but more than or equal to 5
    (df['Experience'] >= 10)  # Senior: 10 or more years of experience
]

names = ['Junior', 'Mid', 'Senior']

# Create the new 'Experience Level' column
df['Experience Level'] = np.select(conditions, names, default='Unknown')

# Check the result
df.head()
```

```python
# 18. Add a new row with specific values using the append() method. 
new_row = {
    'ID': 6,
    'Name': 'John Doe',
    'Age': 30,
    'Gender': 'Male',
    'Department': 'Sales',
    'Salary': 5500,
    'Joining Date': '2023-01-15',
    'Performance Score': 4.5,
    'Experience': 7,
    'Status': 'Active',
    'Location': 'Chicago',
    'Session': 'Morning'
}
df = df.append(new_row, ignore_index=True)
df
```

```python
# 19. Delete the column Session from the DataFrame. 
df_drop = df.drop('Session', axis=1, inplace=True)
```

```python
# 20. Delete rows where the Age is greater than 60. 
indices_to_drop = df[df['Age'] > 60].index
df = df.drop(indices_to_drop)
df.head()
```

```python
# 21. Add a new column Age Group with values "Young", "Middle-aged", or "Senior" based on Age.
conditions = [
    (df['Age'] < 30),
    (df['Age'] < 60),
    (df['Age'] > 60)
]

names = ['Young', 'Middle-Aged', 'Senior']

df['Age Group'] = np.select(conditions, names, default='Unknown')
df.head()
```

```python
# 22. Identify all columns with missing values using the isna() method.
df.columns[df.isna().any()]
```

```python
# 23. Fill the missing values in Performance Score with the average score of that column. 
avg_score =  df['Performance Score'].mean()

df['Performance Score'] = df['Performance Score'].fillna(avg_score)

df.head()
```

```python
# 24. Drop all rows with missing values using the dropna() method. 
df.dropna(axis=0)
```

```python
# 25. Replace missing values in Performance Score with the median of the scores, grouped by Department. 
df['Performance Score'] = df.groupby('Department')['Performance Score'].transform(
    lambda x: x.fillna(x.median())
)

df.head(20)
```

```python
#  48. Create a DataFrame with employee ID and Project Count, then merge it with the original dataset. Calculate the average Salary per project

data2 = {
    'ID': [356, 388, 403, 539, 829, 733, 798, 804, 880, 988],
    'Project Count': [3, 2, 5, 4, 3, 1, 2, 3, 2, 4]
}
df2 = pd.DataFrame(data2)
merged_df = pd.merge(df, df2, on='ID', how='inner')
merged_df['Avg Salary Per Project'] = merged_df['Salary'] / merged_df['Project Count']
```

```python
# 49. Calculate the percentage increase in Salary for each employee compared to the average Salary in their Department. 

avg_salary_by_dept = df.groupby('Department')['Salary'].mean().reset_index()
avg_salary_by_dept.rename(columns={'Salary': 'Avg Salary'}, inplace=True)

# Step 3: Merge the average salary back with the original dataset
df = pd.merge(df, avg_salary_by_dept, on='Department', how='left')

# Step 4: Calculate percentage increase
df['% Increase Compared to Dept Avg'] = ((df['Salary'] - df['Avg Salary']) / df['Avg Salary']) * 100
```
